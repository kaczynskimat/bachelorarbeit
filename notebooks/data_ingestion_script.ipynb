```python
import glob
import pandas as pd
import os
```

```python
# Data Ingestion Script for London Households Energy Dataset

# Please note that the data is not available on GitHub due to GitHub's file size limitations.
# The data can be downloaded here: https://data.london.gov.uk/dataset/smartmeter-energy-consumption-data-in-london-households-vqm0d/
path = r"PLACEHOLDER_FOR_YOUR_PATH" # Update this path to the local directory containing the raw csv files
# Get data file names
all_files = glob.glob(os.path.join(path, "*.csv"))
# start and end date for the interval for analysis
start_date = '2012-11-05 00:00:00'
end_date = '2012-11-12 00:00:00'
# output file
output_file = 'output.csv'
```

```python
add_header = True
for file in all_files: 
    df = pd.read_csv(file)
    df['DateTime'] = pd.to_datetime(df['DateTime'])
    mask = (df['DateTime'] >= start_date) & (df['DateTime'] < end_date)
    df2 = df.loc[mask]

    df2.to_csv(output_file, mode="a", index=False, header=add_header)
    add_header = False
print(f"Finished.") 
```
